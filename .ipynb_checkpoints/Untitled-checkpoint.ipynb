```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("TACs.xlsx", sheet_name=None)
relevant_keys = ['PT 300 ml Qclear 2', 'PT 300 ml Qclear 1', 'PT 200 ml Qclear 2', 'PT 200 ml Qclear 1', 'PT 100 ml Qclear 2', 'PT 100 ml Qclear 1']
qclear = [2, 1, 2, 1, 2, 1]
sizes = [300, 300, 200, 200, 100, 100]
```

```python
frames = []
for key, clear, size in zip(relevant_keys, qclear, sizes):
    print(clear, size)
    frame = data[key]
    df = frame.loc[5:]
    df.columns = ["time", "input", "tissue"]
    df[['time_start', 'time_end']] = df['time'].str.split(' - ', 1, expand=True)
    df["qclear"] = clear
    df["size"] = size
    df = df.astype({'input': 'float64', 'tissue': 'float64', 'time_start': 'int32', 'time_end': 'int32'})
    df = df.drop(["time"], axis=1)
    frames.append(df)
df = pd.concat(frames)
```

```python
df.groupby(['qclear', 'size']).size()
```

```python
for name, group in df.groupby(['qclear', 'size']):
    print(f"QCLEAR: {group['qclear'].iloc[1]}, SIZE: {group['size'].iloc[1]} ml")
    plt.plot(group["time_start"], group["tissue"], color="b")
    plt.plot(group["time_start"], group["input"], color="r")
    plt.show()
```
